```python
import re, string, nltk, spacy,emoji,contractions
```

```python
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, SnowballStemmer
from nltk.tokenize import word_tokenize,sent_tokenize
from spacy.lang.en.stop_words import STOP_WORDS as spacy_stopwords
```

```python
text = text.lower()
```

```python
text = contractions.fix(text) #contractions convert - let's into let us kind of
```

```python
text = emoji.replace_emoji(text, replace=' ') #to replace emoji
```

```python
text =re.sub(r" #\w+","",text)   #remove the '#' or whatever you want to remove
```

```python
texts = text.translate(str.maketrans("","",string.punctuation))  #to remove all punctuations
```

```python
text = sent_tokenize(text=texts)
```
